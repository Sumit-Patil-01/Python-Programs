```python
# Printing double quotes with a string variable
my_text = "Python"

# Method 1: Using single quotes to define the string
print(f'She said, "{my_text} is fun!"')

# Method 2: Using escape character within double quotes
print(f"He replied, \"{my_text} is powerful.\"")

# Method 3: Using f-string with mixed quotes (Python 3.6+)
print(f"""The word is "{my_text}"""")
```

```python
# Ignoring escape sequences in the string using raw strings

# Without raw string, \n is interpreted as a newline
path_normal = "C:\nUsers\Documents\file.txt"
print(f"Normal string with escape: {path_normal}")

# With raw string (r-prefix), \n is treated as literal characters
path_raw = r"C:\nUsers\Documents\file.txt"
print(f"Raw string ignoring escape: {path_raw}")

# Another example: backslashes are often seen in regular expressions
regex_pattern = r"\d+\s\w+"
print(f"Raw string for regex: {regex_pattern}")
```
